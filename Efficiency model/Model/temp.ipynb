```python
from module import X,Y
import numpy as np
import torch
```

```python
def sigmoid(z):
    z = np.clip(z, -500, 500) 
    return 1/(1+np.exp(-z))

def predict(X, W, b):
    z = np.dot(X, W) + b
    return sigmoid(z)  


def bce(y_pred, Y):
    m = Y.shape[0]
    eps = 1e-7
    return -(1/m) * np.sum(Y*np.log(y_pred+eps) + (1-Y)*np.log(1-y_pred+eps))


def predict_class(X, W, b, threshold=0.5):
    probs = predict(X, W, b)
    return (probs >= threshold).astype(int)

```

```python
def gradient(y_pred):
    m=X.shape[0]
    dW = (1/m) * np.dot(X.T, (y_pred - Y))   # (n, 1)
    db = (1/m) * np.sum(y_pred - Y)    
    return dW,db
Y.shape[0]
```

```python
X=X.numpy()
Y=Y.numpy()
```

```python
# parameter initializing 

X = (X - X.mean(axis=0)) / X.std(axis=0)
W = np.random.randn(X.shape[1], 1) * 0.01
b = np.zeros((1, 1))
lr=0.03
epochs=1000
```

```python
for epoch in range(epochs):
    y_pred=predict(X,W,b)
    loss=bce(y_pred,Y)
    dw,db=gradient(y_pred)
    W-=lr*dw
    b-=lr*db
    
    if epoch%10==0:
        y_class = predict_class(X, W, b)
        acc = np.mean(y_class.flatten() == Y.flatten()) * 100
        print(f"Epoch {epoch} | Loss: {loss:.4f} | Accuracy: {acc:.2f}%")

```
